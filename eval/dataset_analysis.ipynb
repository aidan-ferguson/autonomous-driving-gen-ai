```python
import os
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
```

```python
class_str_mapping = {
    0: "blue",
    1: "orange",
    2: "large_orange",
    3: "yellow "
}


def dataset_statistics(dataset_dir):
    labels_dir = os.path.join(dataset_dir, "labels")
    labels = []
    for label_file in os.listdir(labels_dir):
        with open(os.path.join(labels_dir, label_file), "r") as file:
            data = file.readlines()
            data = [int(row.split(" ")[0]) for row in data]
        labels.extend(data)

    plt.hist(labels)
    plt.show()
    print(Counter([class_str_mapping[label] for label in labels]))
    print(np.sum(labels))
```
